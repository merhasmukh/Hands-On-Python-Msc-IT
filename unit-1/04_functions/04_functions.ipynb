# Functions and Functional Tools in Python

## 📖 1. Defining Functions

Functions allow you to reuse code. You define them with the `def` keyword.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))       # Output: Hello, Alice!

def divide_and_remainder(a, b):
    return a // b, a % b  # Returns a tuple

q, r = divide_and_remainder(10, 3)
print(f"Quotient: {q}, Remainder: {r}") # Output: Quotient: 3, Remainder: 1
```

## 📖 2. Default and Keyword Arguments

You can provide default values for parameters. When calling functions, you can pass arguments by position or by keyword.

```python
def make_coffee(size="Medium", sugar=2):
    print(f"Making a {size} coffee with {sugar} sugars.")

make_coffee()                          # Uses defaults
make_coffee("Large")                   # Positional override
make_coffee(sugar=0, size="Small")     # Keyword override (order doesn't matter)
```

## 📖 3. *args and **kwargs

`*args` collects extra positional arguments into a tuple.
`**kwargs` collects extra keyword arguments into a dictionary.

```python
def my_sum(*args):
    total = 0
    for num in args:
        total += num
    return total

print(my_sum(1, 2, 3, 4))    # Output: 10

def user_profile(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

user_profile(name="Alice", age=22, course="MSc IT")
```

## 📖 4. Variable Scope (LEGB)

Local -> Enclosing -> Global -> Built-in

```python
x = 10  # Global

def modify_global():
    global x
    x = 20

modify_global()
print(x)  # Output: 20
```

## 📖 5. Lambda Functions, map(), and filter()

Lambdas are small, anonymous functions. Often used with `map` and `filter`.

```python
add = lambda x, y: x + y
print(add(5, 3))             # Output: 8

nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)               # Output: [1, 4, 9, 16]

evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)                 # Output: [2, 4]
```

## 📖 6. Decorators

Decorators wrap a function to modify its behavior without changing its code.

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_whee():
    print("Whee!")

say_whee()
```

## 📖 7. Generators and Iterators

Generators use `yield` to return a value and suspend their state. They are memory efficient.

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)
# Output: 3, 2, 1
```

## ⚠️ Common Mistakes
- Using a mutable default argument (like `lst=[]` instead of `lst=None`).
- Forgetting the `return` statement (function returns `None` implicitly).
- Recursion without a base case (leads to RecursionError).
