# Problem Solving with Python

## 📖 1. Pattern Programs

Pattern programs help you build logic for nested loops.

```python
n = 5
# Right Triangle
for i in range(1, n + 1):
    print("* " * i)

print("-----")
# Pyramid
for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars = "* " * i
    print(spaces + stars)
```

## 📖 2. Searching Algorithms

- **Linear Search**: Check each element one by one. O(N) complexity.
- **Binary Search**: Requires sorted array. Check the middle, halve the search space. O(log N) complexity.

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

sorted_arr = [1, 3, 5, 7, 9, 11]
print("Index:", binary_search(sorted_arr, 7)) # Output: Index: 3
```

## 📖 3. Sorting Algorithms

- **Bubble Sort**: repeatedly steps through the list, compares adjacent elements and swaps them if they are in the wrong order. O(N^2) complexity.
- Python's built-in `sort()` and `sorted()` use **Timsort**, which is highly optimized (O(N log N)).

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted:", bubble_sort(arr)) # Output: Sorted: [11, 12, 22, 25, 34, 64, 90]
```

## 📖 4. Basic Algorithmic Thinking & Complexity

**Time Complexity Intuition:**
- `O(1)`: Accessing an array by index. (Instant)
- `O(log N)`: Binary search. (Extremely fast, even for billions of items)
- `O(N)`: Linear search, single loop. (Fast)
- `O(N^2)`: Nested loops, Bubble sort. (Slow for large N)
- `O(2^N)`: Recursive Fibonacci without memoization. (Terrible for large N)

## 📖 5. Classic Problems

Here are a few classic interview/practice problems.

```python
# FizzBuzz
def fizzbuzz(n):
    for i in range(1, n + 1):
        if i % 15 == 0: print("FizzBuzz", end=" ")
        elif i % 3 == 0: print("Fizz", end=" ")
        elif i % 5 == 0: print("Buzz", end=" ")
        else: print(i, end=" ")
    print()

fizzbuzz(15)

# Palindrome Check
def is_palindrome(s):
    return s == s[::-1]

print("racecar is palindrome?", is_palindrome("racecar")) # Output: True
```

## ⚠️ Common Mistakes
- Off-by-one errors in nested loops (e.g. `n-i` vs `n-i-1`).
- Modifying a list while you are iterating over it.
- Forgetting the base case in a recursive function.
