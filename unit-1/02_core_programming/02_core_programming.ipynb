# Core Programming in Python

## 📖 1. Variables, Constants, and Data Types

In Python, variables are created the moment you first assign a value to them. You don't need to declare their type. 
There are no true "constants" in Python, but by convention, we use ALL_CAPS for variables that shouldn't change.

```python
# 1. Variables and Data Types
age = 25                  # int
height = 1.75             # float
name = "Alice"            # str
is_student = True         # bool
nothing = None            # NoneType

# Constants (convention only)
PI = 3.14159

print(name, "is", age, "years old.")
# Output: Alice is 25 years old.
```

## 📖 2. Type Checking and Type Casting

You can check the type of a variable using `type()`. You can convert (cast) between types using functions like `int()`, `float()`, `str()`, and `bool()`.

```python
print(type(age))          # Output: <class 'int'>

# Type Casting
num_str = "100"
num_int = int(num_str)
print(num_int + 50)       # Output: 150

bool_val = bool(0)        # 0 is False, any other number is True
print(bool_val)           # Output: False
```

## 📖 4. String Formatting

Python 3.6+ introduced **f-strings**, which are the preferred way to format strings.

```python
course = "Python"
version = 3.11

# f-string (Recommended)
print(f"Welcome to {course} {version}!")  # Output: Welcome to Python 3.11!

# .format() method
print("Welcome to {} {}!".format(course, version))

# Old % style
print("Welcome to %s %.2f!" % (course, version))
```

## 📖 5. Operators

Python has various operators: arithmetic, comparison, logical, assignment, identity, and membership.

```python
a = 10
b = 3

print("a // b =", a // b)  # Floor division (Output: 3)
print("a % b =", a % b)    # Modulo (Output: 1)
print("a ** b =", a ** b)  # Exponentiation (Output: 1000)

# Logical
print(True and False)      # Output: False

# Identity
x = [1, 2, 3]
y = [1, 2, 3]
print(x is y)              # Output: False (different objects in memory)

# Membership
print(2 in x)              # Output: True
```

## 📖 6. Decision Making

Control the flow of your program with `if`, `elif`, and `else`.

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")
# Output: Grade: B

# Ternary Operator
result = "Pass" if score >= 50 else "Fail"
print(result) # Output: Pass
```

## 📖 7. Loops

Use `for` loops to iterate over a sequence. Use `while` loops to repeat as long as a condition is true.

```python
# For loop
for i in range(3):
    print(f"For loop iteration: {i}")

# While loop
count = 0
while count < 2:
    print(f"While loop count: {count}")
    count += 1
```

## 📖 8. Loop Control

- `break`: Exits the loop entirely
- `continue`: Skips to the next iteration
- `pass`: Does nothing (acts as a placeholder)

```python
for i in range(5):
    if i == 2:
        continue  # Skip 2
    if i == 4:
        break     # Stop at 4
    print(i)
# Output:
# 0
# 1
# 3
```

## 📖 9. Built-in Helpers

- `range(start, stop, step)`: Generates a sequence of numbers.
- `enumerate()`: Adds a counter to an iterable.
- `zip()`: Combines multiple iterables.

```python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 95]

# enumerate
for idx, name in enumerate(names, start=1):
    print(f"{idx}. {name}")

# zip
for name, score in zip(names, scores):
    print(f"{name} scored {score}")
```

## ⚠️ Common Mistakes
- Forgetting that `input()` returns a string.
- Confusing `=` (assignment) with `==` (comparison).
- Off-by-one errors with `range()` (stop value is exclusive).
