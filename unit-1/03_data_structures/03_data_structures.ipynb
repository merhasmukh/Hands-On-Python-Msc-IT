# Built-in Data Structures in Python

## 📖 1. Strings

Strings are sequences of characters. They are immutable (cannot be changed after creation).

```python
text = "  Python Programming  "

# Common Methods
print(text.strip())          # Output: 'Python Programming'
print(text.lower())          # Output: '  python programming  '
print(text.upper())          # Output: '  PYTHON PROGRAMMING  '
print(text.replace(" ", "-")) # Output: '--Python-Programming--'

words = text.split()
print(words)                 # Output: ['Python', 'Programming']
print("-".join(words))       # Output: 'Python-Programming'

# Indexing and Slicing
s = "Hello World"
print(s[0])                  # Output: 'H'
print(s[0:5])                # Output: 'Hello'
print(s[-5:])                # Output: 'World'
print(s[::-1])               # Output: 'dlroW olleH' (Reverse)
```

## 📖 2. Lists

Lists are ordered, mutable collections. They can contain mixed data types.

```python
fruits = ["apple", "banana", "cherry"]

# Methods
fruits.append("date")
fruits.insert(1, "blueberry")
fruits.remove("banana")
popped = fruits.pop()

print(fruits)                # Output: ['apple', 'blueberry', 'cherry']
print(f"Popped: {popped}")   # Output: Popped: date

# Sorting
nums = [5, 2, 9, 1]
nums.sort()
print(nums)                  # Output: [1, 2, 5, 9]
```

## 📖 3. Tuples

Tuples are ordered, immutable collections. Use them when data should not change.

```python
point = (10, 20)

# Unpacking
x, y = point
print(f"X: {x}, Y: {y}")     # Output: X: 10, Y: 20

# Tuples are faster and safer than lists for fixed data.
```

## 📖 4. Sets

Sets are unordered collections of unique elements. Great for finding unique items and mathematical operations.

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

s1.add(5)
s1.remove(1)

print(s1 | s2)               # Union: {2, 3, 4, 5, 6}
print(s1 & s2)               # Intersection: {3, 4, 5}
print(s1 - s2)               # Difference: {2}

# Remove duplicates from list
dupes = [1, 2, 2, 3, 3, 3]
print(list(set(dupes)))      # Output: [1, 2, 3]
```

## 📖 5. Dictionaries

Dictionaries store key-value pairs. Keys must be unique and immutable.

```python
student = {
    "name": "Alice",
    "age": 22,
    "courses": ["Python", "Data Science"]
}

print(student["name"])       # Output: Alice
print(student.get("grade", "N/A")) # Output: N/A

student["age"] = 23
student.update({"city": "Mumbai"})

for k, v in student.items():
    print(f"{k}: {v}")
```

## 📖 6. Comprehensions

Comprehensions are concise ways to create lists, dictionaries, or sets.

```python
# List Comprehension
squares = [x**2 for x in range(5)]
print(squares)               # Output: [0, 1, 4, 9, 16]

evens = [x for x in range(10) if x % 2 == 0]
print(evens)                 # Output: [0, 2, 4, 6, 8]

# Dictionary Comprehension
sq_dict = {x: x**2 for x in range(5)}
print(sq_dict)               # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

## 📖 7. Nested Data Structures

You can nest data structures indefinitely.

```python
classroom = [
    {"name": "Alice", "grades": [90, 85]},
    {"name": "Bob", "grades": [70, 75]}
]

print(classroom[0]["grades"][1]) # Output: 85
```

## ⚠️ Common Mistakes
- Mutating a list while iterating over it.
- Trying to use mutable types (like lists) as dictionary keys.
- Using string concatenation (+) in loops instead of `.join()`.
