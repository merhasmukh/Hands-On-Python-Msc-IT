# Modules, Packages, and Error Handling

## 📖 1. Modules

A module is just a Python file (`.py`) containing code. You can use it in another file using the `import` statement.

```python
import math
print(math.pi)               # Output: 3.141592653589793

from math import sqrt
print(sqrt(16))              # Output: 4.0

import math as m
print(m.pow(2, 3))           # Output: 8.0
```

## 📖 5. Exception Handling

Use `try/except` to catch errors gracefully so your program doesn't crash.

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error caught: {e}")
except TypeError:
    print("Type error!")
else:
    print("Success!") # Runs if NO exception occurs
finally:
    print("Cleaning up...") # ALWAYS runs
```

## 📖 6. Raising Custom Exceptions

You can trigger an exception manually using `raise`. You can also define your own exception classes.

```python
class NegativeAgeError(ValueError):
    pass

def set_age(age):
    if age < 0:
        raise NegativeAgeError("Age cannot be negative!")
    print(f"Age set to {age}")

try:
    set_age(-5)
except NegativeAgeError as e:
    print(e)
```

## 📖 7. Type Hints (Python 3.10+)

Type hints make your code more readable and help IDEs catch errors. Use `|` for Union types.

```python
def process_data(data: list[int] | str) -> int:
    if isinstance(data, str):
        return len(data)
    return sum(data)

print(process_data([1, 2, 3])) # Output: 6
print(process_data("hello"))   # Output: 5
```

## ⚠️ Common Mistakes
- Using a bare `except:` block (it catches everything, including KeyboardInterrupt, making it hard to exit the program).
- Not checking `if __name__ == "__main__":` when writing modules.
