# Object-Oriented Programming (OOP) in Python

## 📖 1. What is OOP?

Object-Oriented Programming is a paradigm based on "objects". An object combines data (attributes) and behavior (methods) into a single unit.

- **Class**: A blueprint for creating objects (e.g., `Car`).
- **Object**: An instance of a class (e.g., `my_honda_civic`).

## 📖 2. Defining a Class and Using `__init__`

The `__init__` method is the constructor. It initializes the object's attributes when it's created. `self` refers to the specific object being created.

```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance attributes (unique to each instance)
        self.name = name
        self.age = age

    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"

# Creating objects (instances)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

print(dog1.name)             # Output: Buddy
print(dog1.bark())           # Output: Buddy says Woof!
print(dog2.species)          # Output: Canis familiaris
```

## 📖 3. Encapsulation (Private Attributes)

Encapsulation hides internal state. In Python, prefixing an attribute with an underscore `_` means "protected", and double underscore `__` means "private" (name mangling).

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        return "Invalid deposit amount."

    # Getter using @property decorator
    @property
    def balance(self):
        return self.__balance

account = BankAccount("Alice", 100)
print(account.deposit(50))   # Output: Deposited 50. New balance: 150
print(account.balance)       # Output: 150
# print(account.__balance)   # This would raise an AttributeError!
```

## 📖 4. Inheritance

Inheritance allows a new class to inherit attributes and methods from an existing class. Use `super()` to call the parent class's constructor.

```python
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance, interest_rate):
        super().__init__(owner, balance) # Call parent's __init__
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.deposit(interest)
        return f"Interest added. New balance: {self.balance}"

savings = SavingsAccount("Bob", 1000, 0.05)
print(savings.add_interest()) # Output: Interest added. New balance: 1050.0
```

## 📖 5. Polymorphism

Polymorphism allows different classes to define methods with the same name. You can use objects of different types interchangeably.

```python
class Cat:
    def speak(self):
        return "Meow"

class Duck:
    def speak(self):
        return "Quack"

def animal_sound(animal):
    print(animal.speak())

cat = Cat()
duck = Duck()

animal_sound(cat)            # Output: Meow
animal_sound(duck)           # Output: Quack
```

## 📖 6. Dunder (Magic) Methods

Special methods surrounded by double underscores (e.g., `__str__`, `__len__`, `__add__`) define how objects behave with built-in operations.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(2, 4)
v2 = Vector(3, 1)

print(v1)                    # Output: Vector(2, 4)
v3 = v1 + v2                 # Uses __add__
print(v3)                    # Output: Vector(5, 5)
```

## 📖 7. Class Methods and Static Methods

- `@classmethod`: Takes `cls` as the first argument. Can access and modify class state. Used for alternative constructors.
- `@staticmethod`: Takes neither `self` nor `cls`. Just a regular function that belongs in the class's namespace.

```python
class MathHelper:
    @staticmethod
    def add(a, b):
        return a + b

print(MathHelper.add(5, 10)) # Output: 15
```

## ⚠️ Common Mistakes
- Forgetting `self` as the first parameter in instance methods.
- Modifying a class attribute accidentally via an instance (`self.class_attr = x` creates a shadowing instance attribute!).
- Trying to access a `__private` attribute directly from outside.
