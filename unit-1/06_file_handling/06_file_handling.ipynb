# File Handling in Python

## 📖 1. Why File Handling?

Data stored in variables is lost when the program ends. File handling allows us to store data persistently on the hard drive.

## 📖 2. Opening and Reading Files

We use `open()` to get a file object. Modes include `r` (read), `w` (write), `a` (append).

```python
# The manual way (NOT recommended, as you might forget to close)
# f = open("sample.txt", "r")
# content = f.read()
# print(content)
# f.close()
```

## 📖 3. Context Managers (`with` statement)

The `with` statement automatically closes the file for you, even if an exception occurs! This is the Pythonic way.

```python
try:
    with open("sample.txt", "r") as file:
        # read() reads the entire file
        # readline() reads a single line
        # readlines() returns a list of lines
        for line in file:
            print(line.strip()) # strip() removes the trailing newline
except FileNotFoundError:
    print("File not found!")
```

## 📖 4. Writing Files

Use mode `w` to overwrite or create a file, and `a` to append.

```python
with open("output.txt", "w") as file:
    file.write("Hello, World!\n")
    file.writelines(["Line 2\n", "Line 3\n"])
```

## 📖 5. CSV Files

CSV (Comma Separated Values) is a common format for tabular data. Python has a built-in `csv` module.

```python
import csv

# Reading a CSV as lists
with open("sample.csv", "r") as f:
    reader = csv.reader(f)
    header = next(reader) # Skip the header row
    for row in list(reader)[:2]: # Print first 2 rows
        print(row)

# Reading a CSV as dictionaries (Better!)
with open("sample.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in list(reader)[:2]:
        print(row["Name"], "scored", row["Score"])
```

## 📖 6. JSON Files

JSON (JavaScript Object Notation) is heavily used in web APIs. It maps directly to Python dictionaries.

```python
import json

# Reading JSON
with open("sample.json", "r") as f:
    data = json.load(f)
    print("University:", data["university"])
    print("First student:", data["students"][0]["name"])

# Writing JSON
out_data = {"status": "success", "code": 200}
with open("output.json", "w") as f:
    json.dump(out_data, f, indent=4)
```

## 📖 7. File Paths (pathlib)

The modern way to handle file paths is using the `pathlib` module.

```python
from pathlib import Path

p = Path("sample.txt")
if p.exists():
    print(f"{p.name} exists!")
    print(f"Absolute path: {p.absolute()}")
```

## ⚠️ Common Mistakes
- Forgetting to close the file (always use `with`!).
- Opening a file in `w` mode when you mean to read (it will erase the file!).
- Not handling `FileNotFoundError`.
