# Continuous Integration and Continuous Deployment (CI/CD)

## 📖 1. What is CI/CD?

- **Continuous Integration (CI)**: Automatically testing your code every time you push it to a repository. This ensures that new code doesn't break old features.
- **Continuous Deployment (CD)**: Automatically pushing the code to the production server if all the tests pass.

## 📖 2. GitHub Actions

GitHub provides a free built-in tool for CI/CD called **GitHub Actions**. 
You configure it by creating YAML files inside a specific folder in your repository: `.github/workflows/`.

## 📖 3. Anatomy of a GitHub Actions Workflow

A workflow file (e.g. `main.yml`) has three main parts:

1. **Events (on)**: When should this workflow run? (e.g., `on: push` or `on: pull_request`).
2. **Jobs**: What broad task are we doing? (e.g., `test` or `deploy`). Jobs can run in parallel.
3. **Steps**: The individual commands to run inside the job (e.g., `pip install -r requirements.txt`, `pytest`).

```python
# Example of what the steps look like in YAML:
# 
# steps:
# - uses: actions/checkout@v3  # 1. Download the code to the GitHub server
# - name: Set up Python
#   uses: actions/setup-python@v4
#   with:
#     python-version: '3.11'   # 2. Install Python 3.11
# - name: Install dependencies
#   run: pip install pytest    # 3. Install pytest
# - name: Run tests
#   run: pytest                # 4. Run the tests!
```

## 📖 4. Secrets in CI/CD

If your tests require a database password or an API key, you cannot put them in the YAML file. 
Instead, you go to the **Settings -> Secrets** tab in your GitHub repository and add them there. 
You can access them in the YAML file using `${{ secrets.MY_PASSWORD }}`.

## ⚠️ Common Mistakes
- Putting the workflow file in the wrong folder (it MUST be exactly `.github/workflows/`).
- YAML indentation errors (YAML is extremely strict about spaces. Use 2 spaces for indentation!).
- Hardcoding passwords in the workflow file.
