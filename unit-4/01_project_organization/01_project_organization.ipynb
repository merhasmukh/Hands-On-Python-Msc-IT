# Project Organization & Configuration

## 📖 1. The Twelve-Factor App

The **Twelve-Factor App** methodology is a set of best practices for building modern, scalable, and maintainable software-as-a-service (SaaS) applications.

Two critical factors we focus on here are:
1. **Codebase**: One codebase tracked in version control, many deploys (Development, Staging, Production).
2. **Config**: Store config in the environment, NOT in the code.

## 📖 2. Handling Secrets and Configuration

Never hardcode sensitive information (passwords, API keys, database URIs) directly into your `.py` files. If you push your code to GitHub, anyone can see them!

Instead, we use **Environment Variables**.

````python
import os

# BAD (Hardcoded):
DB_PASSWORD = "super_secret_p@ssw0rd"

# GOOD (From Environment):
DB_PASSWORD = os.getenv("DB_PASSWORD")
print(f"The password is loaded from the environment (currently: {DB_PASSWORD})")
````

## 📖 4. The `.env.example` File

You **MUST NEVER** commit your `.env` file to version control. It should be added to your `.gitignore`.

However, other developers need to know what environment variables your app requires. 
So, we create a `.env.example` file that IS committed to Git, but with fake or blank values.

```ini
# .env.example
DB_USER=your_db_username_here
DB_PASS=your_db_password_here
```

## ⚠️ Common Mistakes
- Hardcoding AWS keys or Database passwords in `app.py`.
- Forgetting to add `.env` to `.gitignore` before committing.
- Expecting `os.getenv("PORT")` to return an integer (it always returns a string; you must cast it: `int(os.getenv("PORT"))`).
