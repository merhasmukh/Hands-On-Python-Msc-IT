# Deployment Fundamentals

## 📖 1. WSGI vs ASGI

To deploy a Python web application, you need an application server that translates HTTP requests into Python function calls.

- **WSGI (Web Server Gateway Interface)**: The synchronous standard. Used by **Flask** and **Django**. The most popular server is **Gunicorn**.
- **ASGI (Asynchronous Server Gateway Interface)**: The modern, asynchronous standard. Used by **FastAPI**. The most popular server is **Uvicorn**.

## 📖 2. Gunicorn + Uvicorn Workers

Uvicorn is great for handling async requests, but it is a single-process server. 
In production, we use **Gunicorn** as a process manager to run multiple *workers*, and we tell it to use Uvicorn as the worker class!

**Command:**
```bash
gunicorn main:app -w 4 -k uvicorn.workers.UvicornWorker
```
- `-w 4`: Start 4 worker processes (usually calculated as `(2 x $num_cores) + 1`).
- `-k`: Use the Uvicorn worker class.

## 📖 3. Reverse Proxy with Nginx

Even with Gunicorn running, you do NOT expose it directly to the internet (port 80/443). 
Instead, you put a **Reverse Proxy** (like **Nginx**) in front of it.

**Why Nginx?**
1. It handles static files (images, CSS) much faster than Python.
2. It provides SSL/TLS termination (HTTPS).
3. It provides load balancing and caching.
4. It acts as a security buffer.

## 📖 4. HTTPS and SSL/TLS

HTTP data is sent in plain text. If someone intercepts the traffic on a public Wi-Fi network, they can see passwords and tokens.

**HTTPS** encrypts the data using SSL/TLS certificates.
Today, you can get free SSL certificates using **Let's Encrypt** and a tool called **Certbot**, which automatically configures your Nginx server for HTTPS.

## 📖 5. Application Architecture

A standard production deployment looks like this:

1. **Internet** -> (Port 443/HTTPS)
2. **Nginx** (Terminates SSL, serves static files)
3. -> `proxy_pass` to `http://127.0.0.1:8000`
4. **Gunicorn** (Manages 4 processes)
5. **Uvicorn Workers** (Runs the actual FastAPI code)
6. **FastAPI Application**

## ⚠️ Common Mistakes
- Running `uvicorn main:app` directly on port 80 in production.
- Forgetting to set up a Reverse Proxy.
- Running Flask with the built-in `app.run()` in production instead of WSGI/Gunicorn.
