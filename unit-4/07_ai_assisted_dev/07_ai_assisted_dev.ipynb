# AI-Assisted Development

## 📖 1. What is Prompt Engineering?

AI models like GitHub Copilot, ChatGPT, or Gemini are incredibly powerful coding assistants, but they are only as good as the instructions (prompts) you give them. 

**Prompt Engineering** is the skill of writing clear, highly specific instructions to get the exact code or explanation you need from an AI.

## 📖 2. The 3 Pillars of a Great Coding Prompt

1. **Context**: What are you building? (e.g., "I am building a FastAPI backend using Python 3.11").
2. **Specific Action**: What exactly do you want the AI to do? (e.g., "Write a Pydantic schema for a User that requires an email and a password of at least 8 characters").
3. **Format/Constraints**: How do you want the output? (e.g., "Return ONLY the Python code, with inline comments explaining the validation logic").

## 📖 3. Using AI for Debugging

When you hit an error, do NOT just paste the error code and say "Fix this". 

**Good Debugging Prompt**:
> "I am getting a `sqlalchemy.exc.IntegrityError: UNIQUE constraint failed: users.email` in my FastAPI application when trying to register a user. 
> Here is my route code: [Paste Code]
> How can I catch this exception and return a clean HTTP 400 response to the client instead of crashing?"

## 📖 4. Using AI for Documentation & Tests

Writing tests and documentation is often tedious. This is where AI shines!

**Good Documentation Prompt**:
> "Add Google-style docstrings to this Python class explaining the parameters, return types, and potential exceptions."

**Good Testing Prompt**:
> "Write 3 pytest functions using FastAPI's TestClient to test the `POST /items/` route. Test a successful creation, a missing name field, and an invalid price field."

## 📖 5. Responsible and Ethical AI Use

- **Security**: NEVER paste proprietary company code, API keys, database passwords, or customer data into a public AI chat interface.
- **Understanding**: Do not copy-paste code you do not understand. If you push AI-generated code, YOU are responsible for its bugs and security vulnerabilities.
- **Plagiarism**: Use AI as a pair-programmer, not to do your university assignments for you without learning the concepts!

## ⚠️ Common Mistakes
- Vague prompts like "Make a website using python".
- Blindly trusting AI (AI models can hallucinate and confidently provide code that uses deprecated libraries or introduces security flaws).
- Pasting your AWS Secret Key into ChatGPT.
