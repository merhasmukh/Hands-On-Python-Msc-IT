# Linux Essentials for DevOps

## 📖 1. Why Linux?

Most web servers in the world run Linux (specifically distributions like Ubuntu, Debian, or Alpine). Even if you develop on Windows or macOS, you will deploy to a Linux server or a Linux-based Docker container. Knowing the basics of the Linux command line is mandatory for modern backend development.

## 📖 2. File System Navigation

Here are the absolute basics for moving around the server:

- `pwd`: Print Working Directory (Where am I?)
- `ls`: List files in the current directory (`ls -la` shows hidden files and permissions).
- `cd`: Change Directory (e.g., `cd /var/www/app`).
- `mkdir`: Make Directory (e.g., `mkdir my_project`).
- `rm`: Remove file (`rm -rf folder_name` forcefully removes a directory and everything inside it — use with extreme caution!).

## 📖 3. Reading and Editing Files

- `cat <file>`: Prints the entire contents of a file to the screen.
- `less <file>`: Opens a file for reading, allowing you to scroll up and down (press `q` to quit).
- `tail -f <file>`: Follows the end of a file. Very useful for watching live logs!
- `nano <file>`: A simple text editor for editing files directly in the terminal.

## 📖 4. Permissions (`chmod`)

In Linux, every file has permissions for the Owner, Group, and Others. The permissions are Read (r), Write (w), and Execute (x).

If you write a bash script (`script.sh`) and try to run it (`./script.sh`), you might get a "Permission denied" error.
You fix this by adding the execute permission:
```bash
chmod +x script.sh
```

## 📖 5. Process Management

When a program is running (like your FastAPI server), it is a "process".

- `top` or `htop`: View running processes and CPU/RAM usage.
- `ps aux | grep uvicorn`: Find the Process ID (PID) of your running Uvicorn server.
- `kill <PID>`: Gracefully stops a process.
- `kill -9 <PID>`: Forcefully kills a process.

## ⚠️ Common Mistakes
- Running `rm -rf /` (This destroys your entire server. Never do this).
- Editing server files as the `root` user when you shouldn't be.
- Forgetting that Linux is case-sensitive (`App.py` is different from `app.py`).
