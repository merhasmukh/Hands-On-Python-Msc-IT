# Containerization with Docker

## 📖 1. The "It Works on My Machine" Problem

Have you ever written code that works perfectly on your laptop, but when you send it to a friend or deploy it to a server, it crashes?
This happens due to mismatched Python versions, missing environment variables, or different operating systems.

**Docker** solves this by packaging your code, its dependencies, and the entire operating system into a single, standardized unit called a **Container**.

## 📖 2. Images vs Containers

- **Image**: A read-only template (like a blueprint or a class in OOP). You build an image from a `Dockerfile`.
- **Container**: A running instance of an image (like an object in OOP). You can run many containers from one image.

## 📖 3. The Dockerfile

A `Dockerfile` is a text document that contains all the commands a user could call on the command line to assemble an image.

**Example:**
```dockerfile
# 1. Start from an official Python image
FROM python:3.11-slim

# 2. Set the working directory inside the container
WORKDIR /app

# 3. Copy dependencies and install them
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

# 4. Copy the rest of the application code
COPY . .

# 5. Tell Docker what command to run when the container starts
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "80"]
```

## 📖 4. Basic Docker Commands

- `docker build -t my-app .`: Builds an image named 'my-app' using the Dockerfile in the current directory (`.`).
- `docker run -p 8000:80 my-app`: Runs the container and maps port 8000 on your laptop to port 80 inside the container.
- `docker ps`: Lists all running containers.

## 📖 5. Docker Compose

What if your app needs a Database? You could run two separate containers, but networking them manually is hard.

**Docker Compose** is a tool for defining and running multi-container Docker applications using a YAML file (`docker-compose.yml`).
With one command (`docker compose up`), you can start your API, your Database, and your Nginx proxy all at once!

## ⚠️ Common Mistakes
- Forgetting to map ports using `-p`. If you don't map ports, you cannot access the container from your browser!
- Using `COPY . .` *before* `RUN pip install`. (Docker caches layers. If you copy your code first, Docker will reinstall dependencies every time you change a single line of Python code!).
