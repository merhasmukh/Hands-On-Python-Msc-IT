# Version Control with Git & GitHub

## 📖 1. What is Version Control?

Version control (like Git) tracks changes to your codebase over time. It allows you to:
- Revert to previous versions if you break something.
- Collaborate with other developers without overwriting their code.
- Keep a history of *why* changes were made.

## 📖 2. Basic Git Workflow

Here is the daily lifecycle of a developer using Git:

1. `git status`: Check what files have changed.
2. `git add <file>`: Stage a file (tell Git you want to include this change in the next snapshot).
3. `git commit -m "Message"`: Take the snapshot and attach a descriptive message.
4. `git push`: Send your local commits to a remote server (like GitHub).

## 📖 3. Branching and Pull Requests

Never write code directly on the `main` or `master` branch in a professional setting! 

**Branching**:
```bash
# Create and switch to a new branch called 'feature-login'
git checkout -b feature-login
```
Now you can make changes safely without affecting the main code.

**Pull Requests (PR)**:
Once you finish your feature and `git push` your branch to GitHub, you open a Pull Request. This allows other developers to review your code before it is merged into the `main` branch.

## 📖 4. Commit Hygiene

Write clear, descriptive commit messages.

- **Bad**: `git commit -m "fixed stuff"`
- **Bad**: `git commit -m "updates"`
- **Good**: `git commit -m "Add JWT authentication to login route"`
- **Good**: `git commit -m "Fix typo in database connection string"`

## 📖 5. The `.gitignore` file

Not everything belongs in version control! You should NOT commit:
- Passwords / `.env` files
- Virtual environments (`.venv/`)
- Compiled Python files (`__pycache__/`)
- Database files (`*.sqlite3`)

You tell Git to ignore these files by putting their names in a file called `.gitignore`.

## ⚠️ Common Mistakes
- Running `git add .` without checking `git status` first, accidentally committing secret files.
- Forgetting to pull the latest changes (`git pull`) before starting work, leading to merge conflicts.
- Committing a `.env` file (if you do this, you must change all your passwords immediately!).
