# The Data Layer (FastAPI + SQLAlchemy)

## 📖 1. Standard Project Structure

When integrating a database, we do NOT write everything in `main.py`. A standard FastAPI project is split into several files:

- `database.py`: Connects to the database and creates the `SessionLocal` class.
- `models.py`: Contains the SQLAlchemy ORM models (the Database Tables).
- `schemas.py`: Contains the Pydantic models (Data Validation for API requests/responses).
- `crud.py` (optional): Contains reusable functions to interact with the database.
- `main.py`: Contains the FastAPI app and the routes.

## 📖 2. Models vs Schemas

This is the most confusing part for beginners! We have TWO types of "models".

1. **SQLAlchemy Models** (`models.py`): These represent how data is stored in the **Database**. They use `Column`, `Integer`, `String`.
2. **Pydantic Schemas** (`schemas.py`): These represent how data flows in and out of your **API**. They use standard Python types like `int`, `str`.

## ⚠️ Common Mistakes
- Trying to return a SQLAlchemy object without `from_attributes = True` in the Pydantic schema's config.
- Forgetting to commit to the database (`db.commit()`) when inserting or updating data.
- Confusing Pydantic models (Schemas) with SQLAlchemy models (Database).
