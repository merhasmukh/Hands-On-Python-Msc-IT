# Building APIs & Automatic Documentation

## 📖 1. Path Parameters

You can declare path "parameters" or "variables" with the same syntax used by Python format strings.

```python
# from fastapi import FastAPI
# app = FastAPI()

# @app.get("/items/{item_id}")
# def read_item(item_id: int):  # Note the type hint! 
#     return {"item_id": item_id}
```

Because we added the `int` type hint, FastAPI will automatically **validate** that the input is an integer. If someone visits `/items/foo`, FastAPI will return a friendly JSON error explaining the type mismatch!

## 📖 2. Query Parameters

When you declare function parameters that are NOT part of the path parameters, FastAPI automatically interprets them as "query" parameters (the part of the URL after the `?`).

```python
# @app.get("/users/")
# def read_users(skip: int = 0, limit: int = 10):
#     return {"skip": skip, "limit": limit}

# URL: /users/?skip=5&limit=20
```

## 📖 3. Request Bodies and Pydantic

When you need to send data from a client (let's say, a browser) to your API, you send it as a request body.
To declare a request body, you use **Pydantic** models.

```python
# from pydantic import BaseModel

# class Item(BaseModel):
#     name: str
#     description: str | None = None  # Optional field
#     price: float
#     tax: float | None = None

# @app.post("/items/")
# def create_item(item: Item):
#     # 'item' is now a fully validated Pydantic object!
#     total_price = item.price + (item.tax if item.tax else 0)
#     return {"name": item.name, "total": total_price}
```

## 📖 4. Status Codes and Response Models

You can define the HTTP status code (e.g., 201 Created) directly in the path operation decorator.
You can also define a `response_model` to filter out data (like passwords) before sending it back.

```python
# @app.post("/items/", status_code=201, response_model=Item)
# def create_item(item: Item):
#     return item
```

## 📖 5. Automatic Interactive Documentation

Because you used Python type hints and Pydantic, FastAPI automatically generates documentation for you using the **OpenAPI** standard.

- **Swagger UI**: Go to `http://127.0.0.1:8000/docs` to see an interactive UI where you can call your API directly from the browser.
- **ReDoc**: Go to `http://127.0.0.1:8000/redoc` for an alternative, cleaner reading view.

## ⚠️ Common Mistakes
- Forgetting to inherit from `BaseModel` when creating a Pydantic class.
- Defining a variable in the URL path (e.g. `/users/{id}`) but forgetting to add it as a parameter in the function `def get_user(id: int):`.
