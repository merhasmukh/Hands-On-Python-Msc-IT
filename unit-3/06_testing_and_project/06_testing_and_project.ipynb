# Testing and Mini-Project

## 📖 1. Why Test APIs?

When building APIs, testing manually via Postman or Swagger UI is great for development, but terrible for long-term maintenance. You need automated unit tests to ensure that adding a new feature doesn't break an old one.

**Pytest** is the standard testing framework in Python.

## 📖 3. Project Maintainability

As you've seen in the previous modules, a real FastAPI application should be structured across multiple files. 

A standard structure includes:
- `/app`
  - `main.py`
  - `models.py` (Database)
  - `schemas.py` (Pydantic)
  - `/routers` (Grouped API endpoints)
- `/tests`
  - `test_main.py`
- `requirements.txt`

## ⚠️ Common Mistakes
- Naming your test files incorrectly. Pytest expects files to start with `test_` (e.g. `test_app.py`).
- Naming your test functions incorrectly. Functions must also start with `test_` (e.g. `def test_prediction():`).
- Loading a heavy AI model inside the route function. Always load the model globally when the app starts, so it's ready when the request comes in.
