# Security and Middleware

## 📖 1. Authentication vs Authorization

- **Authentication**: Verifying WHO you are (e.g., logging in with username/password).
- **Authorization**: Verifying WHAT you can do (e.g., are you an admin?).

## 📖 2. OAuth2 Password Bearer and JWT Tokens

FastAPI provides built-in tools for handling OAuth2 with Password Bearer. This is the standard way to handle authentication in modern APIs.

1. The client sends a `username` and `password` to a `/token` endpoint.
2. The server verifies them and returns a **JWT (JSON Web Token)**.
3. For future requests, the client puts this JWT in the HTTP Header: `Authorization: Bearer <token>`.
4. The server validates the token on protected routes.

## ⚠️ Common Mistakes
- Setting `allow_origins=["*"]` while setting `allow_credentials=True`. This is a security risk and browsers will reject it.
- Storing passwords in plain text instead of using a hashing library like `passlib`.
- Putting sensitive data (like passwords) inside the JWT token payload (tokens are Base64 encoded, meaning anyone can read them!).
