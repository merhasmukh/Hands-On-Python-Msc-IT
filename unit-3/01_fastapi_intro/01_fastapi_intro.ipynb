# Introduction to FastAPI

## 📖 1. What is FastAPI?

FastAPI is a modern, fast (high-performance), web framework for building APIs with Python 3.8+ based on standard Python type hints.

**Key Features:**
- **Fast**: Very high performance, on par with NodeJS and Go (thanks to Starlette and Pydantic).
- **Fast to code**: Increase the speed to develop features by about 200% to 300%.
- **Fewer bugs**: Reduce about 40% of human (developer) induced errors.
- **Intuitive**: Great editor support. Completion everywhere.
- **Easy**: Designed to be easy to use and learn.
- **Short**: Minimize code duplication.
- **Robust**: Get production-ready code. With automatic interactive documentation.

## 📖 2. Flask vs FastAPI

| Feature | Flask | FastAPI |
|---------|-------|---------|
| **Style** | Microframework | Microframework (API focused) |
| **Speed** | Fast enough | Very Fast (Async natively) |
| **Data Validation** | Manual / Flask-WTF | Automatic via Pydantic |
| **Documentation** | Manual | Automatic (Swagger/ReDoc) |
| **Concurrency** | Synchronous (WSGI) | Asynchronous (ASGI) |

## 📖 3. Synchronous vs Asynchronous

- **Synchronous (Sync)**: Like a single cook in a kitchen. The cook starts making soup, waits 10 minutes for it to boil, and does NOTHING else until it's done. Then they start chopping vegetables.
- **Asynchronous (Async)**: The cook puts the soup on the stove, and *while waiting* for it to boil, they chop vegetables. They handle multiple tasks concurrently without blocking.

FastAPI uses `ASGI` (Asynchronous Server Gateway Interface) under the hood.

## 📖 5. Uvicorn

FastAPI does not have a built-in development server like Flask's `app.run()`. It relies on an ASGI server called **Uvicorn**.

To run the app above (assuming it's in a file called `main.py`), you run this in your terminal:

```bash
uvicorn main:app --reload
```
- `main`: the file `main.py`.
- `app`: the object created inside of `main.py` with `app = FastAPI()`.
- `--reload`: make the server restart after code changes (for development only).

## ⚠️ Common Mistakes
- Forgetting to use Uvicorn to run the app.
- Confusing `@app.route()` (Flask syntax) with `@app.get()` / `@app.post()` (FastAPI syntax).
- Trying to return raw HTML strings. FastAPI is designed to return JSON. If you return a Python dictionary, FastAPI automatically converts it to JSON!
