## ⚠️ Common Mistakes
- Trying to use `session` without setting `app.secret_key`.
- Forgetting `enctype="multipart/form-data"` on your HTML form when uploading files.
- Trusting user-uploaded filenames. ALWAYS use `werkzeug.utils.secure_filename` before saving.
