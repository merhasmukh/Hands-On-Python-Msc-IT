# Database Fundamentals

## 📖 1. What is a Database Management System (DBMS)?

A database is an organized collection of data. A DBMS is software that interacts with end users, applications, and the database itself to capture and analyze the data.

- **Relational Database (RDBMS)**: Stores data in tables (rows and columns). Examples: MySQL, PostgreSQL, SQLite, Oracle.
- **NoSQL Database**: Stores data in document, key-value, or graph formats. Examples: MongoDB, Redis.

In this unit, we focus on Relational Databases using **SQLite** and **MySQL**.

## 📖 2. SQLite vs MySQL

- **SQLite**: Built into Python! Serverless, lightweight, stores the entire database in a single file. Perfect for development, testing, and small applications.
- **MySQL**: Requires a running server. Robust, handles concurrent users, used for large-scale production applications.

## 📖 3. Connecting to SQLite in Python

Python has a built-in library called `sqlite3`. Let's connect to the `sample.db` file we created for this lesson.

```python
import sqlite3

# Connect to the database (creates the file if it doesn't exist)
conn = sqlite3.connect('sample.db')

# A cursor is used to execute SQL commands
cursor = conn.cursor()

print("Connected to database successfully!")
```

## 📖 4. Basic SQL & CRUD Operations

CRUD stands for **Create, Read, Update, Delete**. These are the four basic functions of persistent storage.

### ✨ CREATE (Insert Data)

```python
# Using parameterized queries (?) prevents SQL Injection attacks!
new_student = ("Eve", 22, "Hyderabad", "B+")

cursor.execute("INSERT INTO students (name, age, city, grade) VALUES (?, ?, ?, ?)", new_student)
conn.commit() # Save changes

print("Row inserted successfully! ID:", cursor.lastrowid)
```

### 📖 READ (Select Data)

```python
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

print("All Students:")
for row in rows:
    print(row) # Output is a tuple: (id, name, age, city, grade)

print("\n--- Filtering ---")
cursor.execute("SELECT name, grade FROM students WHERE grade = ?", ("A+",))
top_students = cursor.fetchall()
print("Top Students:", top_students)
```

### ✏️ UPDATE (Modify Data)

```python
cursor.execute("UPDATE students SET grade = ? WHERE name = ?", ("A", "Bob"))
conn.commit()

print(f"Rows updated: {cursor.rowcount}")
```

### 🗑️ DELETE (Remove Data)

```python
cursor.execute("DELETE FROM students WHERE name = ?", ("Diana",))
conn.commit()

print(f"Rows deleted: {cursor.rowcount}")
```

## 📖 5. Closing the Connection

Always close the connection when you are done to free up resources.

```python
conn.close()
print("Connection closed.")
```

## ⚠️ Common Mistakes
- Forgetting to call `conn.commit()` after an INSERT, UPDATE, or DELETE.
- Using string concatenation `"SELECT * FROM users WHERE name = '" + name + "'"` instead of parameterized queries `?` (This leads to SQL Injection vulnerabilities!).
- Not closing the database connection.
