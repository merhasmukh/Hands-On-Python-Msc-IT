# REST API Fundamentals

## 📖 1. What is an API?

An **API** (Application Programming Interface) allows two applications to talk to each other. 
**REST** (Representational State Transfer) is a set of rules for creating web APIs.

In a REST API, we deal with **resources** (like Users, Posts, Products) and we perform operations on them using **HTTP Methods**.

## 📖 2. HTTP Methods (CRUD mapped to HTTP)

| Operation | HTTP Method | URL Example | Description |
|-----------|-------------|-------------|-------------|
| Create | **POST** | `/api/users` | Create a new user |
| Read | **GET** | `/api/users/1` | Get user with ID 1 |
| Update | **PUT** | `/api/users/1` | Update user with ID 1 |
| Delete | **DELETE** | `/api/users/1` | Delete user with ID 1 |

## 📖 4. Status Codes

It is important to return the correct HTTP Status Code along with your JSON response.

- `200 OK`: Standard success
- `201 Created`: Successfully created a resource (POST)
- `400 Bad Request`: Client sent bad data
- `404 Not Found`: Resource doesn't exist
- `500 Internal Server Error`: Your code crashed

## 📖 6. Testing APIs with Postman

Since browsers only perform `GET` requests from the address bar, we need a tool to test `POST`, `PUT`, and `DELETE` requests.

**Postman** (or Insomnia, or cURL) is the industry standard tool for testing APIs.
1. Open Postman
2. Select the Method (e.g., POST)
3. Enter the URL (e.g., `http://127.0.0.1:5000/api/users`)
4. Go to the `Body` tab -> Select `raw` -> Select `JSON`
5. Type `{"name": "Alice"}` and click **Send**!

## ⚠️ Common Mistakes
- Trying to return a Python dictionary directly from a route (always use `jsonify` or let Flask auto-convert if using Flask 2+).
- Forgetting to set the `Content-Type: application/json` header in Postman when sending POST requests.
- Using a `GET` request when trying to modify data (always use POST/PUT/DELETE for modifications).
